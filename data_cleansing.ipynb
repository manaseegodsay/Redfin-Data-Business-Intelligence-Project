```python
import pandas as pd
import numpy as np
import datetime
```

```python
data = pd.read_csv('data.csv')
```

```python
(row_data, col_data) = data.shape
row_data, col_data
```

```python
#grouping data
def processing_sale_data():
    #s.replace may be alternate
    sale_price = []
    sale_date = []
    commission = []
    states = []
    for index, row in data.iterrows():
        if len(row['SALE PRICE']) >= 5:
            sale_price.append(int(row['SALE PRICE']))
            sale_date.append(row['SALE DATE'])
            commission.append(0.03 * int(row['SALE PRICE']))
            states.append('Sold')
        else:
            sale_price.append(0)
            sale_date.append('-')
            commission.append(0)
            states.append('Pending')
    sale_price = pd.Series(sale_price)
    sale_date = pd.Series(sale_date)
    commission = pd.Series(commission)
    states = pd.Series(states)
    data['STATUS'] = states
    data['SALE PRICE'] = sale_price
    data['SALE DATE'] = sale_date
    data['COMMISSION'] = commission
    
    

processing_sale_data()
data.head(2)
```

```python
def group_data():
    unsold_index = list(data.loc[data['SALE PRICE'] == 0].index.values)
    sold_index = list(data.loc[data['SALE PRICE'] != 0].index.values)
    return unsold_index, sold_index
unsold_index, sold_index = group_data()
data.iloc[sold_index,:].head(3)
```

```python
def add_agent_name():
    agent_name_list = ['-','Adam','Beck','Catherine','Newton','Michael','Kim','Nathan']
    agent_name_series = pd.Series(np.random.randint(1,len(agent_name_list), size=row_data))
    agent_name_series[unsold_index] = 0
    agent_name_series = agent_name_series.map(pd.Series(agent_name_list))
    data['AGENT NAME'] = agent_name_series
add_agent_name()
data.head(3)
```

```python
def add_days_in_market():
    days_series = pd.Series(np.abs(np.random.normal(30,120, size=row_data)))
    days_series[sold_index] = 0
    days_series = days_series.astype(int)
    data['DAYS IN MARKET'] = days_series
add_days_in_market()
data.head(3)
```

```python
def add_price_trend():
    price_trend = pd.DataFrame({'PRICE TREND LAST 1WEEK': np.random.normal(0,10, size=row_data),
                               'PRICE TREND LAST 2WEEK': np.random.normal(0,10, size=row_data),
                               'PRICE TREND LAST 3WEEK': np.random.normal(0,10, size=row_data),
                               'PRICE TREND LAST 4WEEK': np.random.normal(0,10, size=row_data)})
    temp = data['DAYS IN MARKET']
    price_trend['PRICE TREND LAST 1WEEK'].loc[temp <= 7] = 0
    price_trend['PRICE TREND LAST 2WEEK'].loc[temp <= 14] = 0
    price_trend['PRICE TREND LAST 3WEEK'].loc[temp <= 21] = 0
    price_trend['PRICE TREND LAST 4WEEK'].loc[temp <= 28] = 0
    new_data = pd.concat([data, price_trend], axis=1) 
    return new_data
data = add_price_trend()
data
```

```python
data.to_csv('nyc-rolling-sales-processed.csv',sep = ',')
```
